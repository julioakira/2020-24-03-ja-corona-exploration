```python
from Bio.Seq import Seq
import lzma
```

```python
# SARS-CoV-2 genome
with open('genome.txt', 'r') as c:
    c = c.read()
    for s in " \n0123456789":
        c = c.replace(s, '')
```

```python
# SARS-CoV-2 has 8.4kb of genetic data :D
len(lzma.compress(c.encode('utf-8')))
```

```python
def pad_seq(seq):
#     Pad sequence to multiple of 3 with Ns
    return {0: seq, 1: seq+'NN', 2: seq+'N'}[len(seq) % 3]
seq = Seq(pad_seq(c))
# Back to RNA base
seq.transcribe()
```
